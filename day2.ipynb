```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data to plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)

# Plot the first series
plt.plot(x, y1, label='Series 1')

# Shift the second series upwards and plot it
y2_shifted = y2 + y1
plt.plot(x, y2_shifted, label='Series 2')

# Shift the third series upwards and plot it
y3_shifted = y3 + y2_shifted
plt.plot(x, y3_shifted, label='Series 3')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

fig, ax = plt.subplots()

ell = Ellipse(xy=(0,0), width=4, height=8, angle=45, facecolor='red', alpha=0.5)
ax.add_artist(ell)

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 1000)
x = 2 * np.sin(t) * np.sin(t)
y = 2 * np.sin(t) * np.cos(t)

plt.plot(x, y, color='red', linewidth=2)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 1000)
r = 1 + np.cos(theta)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rticks([])
ax.set_yticklabels([])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 1000)
r = 1 - np.cos(theta)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rticks([0.5, 1, 1.5, 2])
ax.set_xticks(np.linspace(0, 2 * np.pi, 8, endpoint=False))
ax.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the chart
labels = ['Apple', 'Banana', 'Mango', 'Grapes']
sizes = [20, 30, 25, 25]

# Plotting the chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, startangle=90, counterclock=False, wedgeprops=dict(width=0.5))

# Cut out the center to make it a donut shape
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.show()
```

```python
# polar pychart

import matplotlib.pyplot as plt
import numpy as np

# create data for the chart
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

# plot the chart
plt.polar(theta, r)
plt.grid(True)

# show the plot
plt.show()
```
